```python
from multiprocessing import Pool
import os
from repository_miner import compute_repo_statistics
```

```python
DATA_PATH = 'C:/Users/AQ38570/Desktop/work/code review and refactoring/data'
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

#change to os.path.join(DATA_PATH,'Openstack','raw data','git repos') for openstack
REPOS_PATH = os.path.join(DATA_PATH,'QT','raw data','git repos')
NB_PRCOCESS = 10
all_results = []
files_lists = os.listdir(REPOS_PATH)
all_files_lists = [os.path.join(REPOS_PATH,file) for file in files_lists]
with Pool(NB_PRCOCESS) as p:
    all_results += p.map(compute_repo_statistics,all_files_lists)
        
```

```python
all_results
```

```python
all_stats = {'commits_count' : 0,'files_count' : 0, 'total_loc': 0 }
for result in all_results: 
    all_stats['commits_count'] += result['commits_count'] 
    all_stats['files_count'] += result['files_count'] 
    all_stats['total_loc'] += result['total_loc'] 
```
